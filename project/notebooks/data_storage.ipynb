```python
# Stage 05 â€” Data Storage 

from pathlib import Path
import pandas as pd

project_root = Path.cwd().parent
RAW_DIR = project_root / "data" / "raw"
PROCESSED_DIR = project_root / "data" / "processed"

# Load raw (KO daily prices)
prices = pd.read_csv(RAW_DIR / "ko_prices.csv", parse_dates=["Date"])

# Every Close cell to numeric
prices["Close"] = pd.to_numeric(prices["Close"], errors="coerce")

# Daily return
df_proc = (
    prices[["Date", "Close"]]
    .rename(columns={"Close": "close"})
    .sort_values("Date")
    .assign(daily_ret=lambda d: d["close"].pct_change())
)

# Save
out_path = PROCESSED_DIR / "ko_daily_returns.parquet"
df_proc.to_parquet(out_path, index=False)

print("Saved:", out_path)
print(df_proc.info())
display(df_proc.head())
display(df_proc.tail())
```
