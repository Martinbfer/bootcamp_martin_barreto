```python
from pathlib import Path
import pandas as pd
import yfinance as yf

project_root = Path.cwd().parent
raw_path = project_root / "data" / "raw"
raw_path.mkdir(parents=True, exist_ok=True)

ticker = "KO"
prices = yf.download(ticker, start="2000-01-01", auto_adjust=True, progress=False).reset_index()
prices.to_csv(raw_path / "ko_prices.csv", index=False)

tkr = yf.Ticker(ticker)
financials = {
    "quarterly_financials": getattr(tkr, "quarterly_financials", pd.DataFrame()).to_dict(),
    "quarterly_balance_sheet": getattr(tkr, "quarterly_balance_sheet", pd.DataFrame()).to_dict(),
    "quarterly_cashflow": getattr(tkr, "quarterly_cashflow", pd.DataFrame()).to_dict(),
    "dividends": getattr(tkr, "dividends", pd.Series()).to_dict(),
}
pd.Series(financials).to_json(raw_path / "ko_financials.json")

print("saved")
display(prices.head())

###Raw data for Coca-Cola (KO) stored in `data/raw/`:
### `coca_cola_prices.csv` — daily adjusted close prices since 2000 (Yahoo Finance).
### `coca_cola_financials.json` — quarterly financial statements (Yahoo Finance).
### API keys managed with `.env` (not committed). Example provided in `.env.example`.
### Validation: prices checked for continuous dates and non-null values; sample rows inspected in notebook. ###
```
